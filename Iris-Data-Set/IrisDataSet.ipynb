# <b>Iris Data Set</b>

The "Iris" dataset was originally published at UCI Machine Learning Repository by Ronald Fisher in 1936. Iris Data Set, this small dataset from 1936 is often used for testing out machine learning algorithms and visualizations (for example, Scatter Plot). Each row of the table represents an iris flower, including its species and dimensions of its botanical parts, sepal and petal, in centimeters. [1] [2]

The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

The data stored was the sepal length in cm, sepal width in cm, petal length in cm, petal width in cm and the class, which consisted of: The Iris Setosa, The Iris Versicolour and The Iris Virginica

# Pandas

pandas is an open source, BSD-licensed library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming language.

### Loading the iris data set

```python
# Import the pandas package.
import pandas as pd

# Loads the iris data set from  URL.
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")

# Show the data
df
```

```python
# Check data type of each of the columns
df.dtypes
```

```python
# Display a list of specific headings
df[['sepal_length', 'sepal_width']]
```

```python
# Displays a specific number of rows between a given range
df[20:30]
```

```python
# Displays a specific row by name and specific number of rows between a given range
df[['petal_length', 'petal_width']][20:40]
```

```python
# Displays the value at a specific given position and given row
df.at[80, 'species']
```

## Graphing the data (With Seaborn and MatPlotLib)

Now that we have this data stored, we can plot it onto a graph. We can do this by using the seaborn package. 

Seaborn is a Python data visualization library based on matplotlib and closely integrated with pandas data structures. It provides a high-level interface for drawing attractive and informative statistical graphics.[3] [4]



```python
# Import seaborn for plotting the data
import seaborn as sea

#Import matplotlib to plot the data
import matplotlib.pyplot as plt

# Plot the species with a '^' symbol to represent the points
graph_plot = sea.pairplot(df, hue='species', markers='^')
```

With this data we can see that the setosa iris stands out significantly moreso than the versicolor and the virginica. 

```python
barPW = sea.barplot(x='species', y='petal_width', data=df)
plt.show()
barPL = sea.barplot(x='species', y='petal_length', data=df)
plt.show()
barSW = sea.barplot(x='species', y='sepal_width', data=df)
plt.show()
barSL = sea.barplot(x='species', y='sepal_length', data=df)
plt.show()
```

With these bar plots we can see even more clearly the different between the three sets of iris flower. The petal length and widths show that the setosa is the shortest iris, with the virginica having the longest petal length and width. But when we look at the sepal width graph we see that the setosa has the highest width, with the versicolor having the shortest width. Although the difference between the largest and shortest is less marginal than the petals length and width.

```python
sea.violinplot(x='species', y='sepal_width', data=df, inner='quartile')
plt.show()
sea.violinplot(x='species', y='sepal_length', data=df, inner='quartile')
plt.show()
sea.violinplot(x='species', y='petal_width', data=df, inner='quartile')
plt.show()
sea.violinplot(x='species', y='petal_length', data=df, inner='quartile')
plt.show()
```

When we plot the lengths and widths on a violin plot then we can clearly see the difference between the species due to the size of the inner quartile ranges.
